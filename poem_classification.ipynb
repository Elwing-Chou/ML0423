```python
import pandas as pd
train_df = pd.read_csv("poem_train.csv", encoding="utf-8")
train_df
```

```python
test_df = pd.read_csv("poem_test.csv", encoding="utf-8")
test_df
```

```python
# 答案: 處理成數字
# replace()
u = train_df["作者"].unique()
trans = {name:i for i, name in enumerate(u)}
reverse_trans = {i:name for name, i in trans.items()}
print(trans)
print(reverse_trans)
y_train = train_df["作者"].replace(trans)
y_test = test_df["作者"].replace(trans)
y_test
```

```python
# 如果詩詞分類, 標點符號和換行, 應該去掉, 因為她不影響結果
import jieba
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。"
s = " ".join(jieba.cut(s))
s.replace("\r", "").replace("\n", "")
```

```python
def poemcut(s):
    s = " ".join(jieba.cut(s))
    s = s.replace("\r", "").replace("\n", "")
    return s
# pandas表格: DataFrame 一欄或者一列: Series
# Series.apply(功能名字)
x_train = train_df["內容"].apply(poemcut)
x_test = test_df["內容"].apply(poemcut)
x_test
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x_train_vec = vec.fit_transform(x_train)
x_test_vec = vec.transform(x_test)
```

```python
# print(x_train_vec)
# vec.vocabulary_
# 檢查。有沒有在裡面
# vec.vocabulary_["。"]
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train_vec, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_vec)
print("正確率:", accuracy_score(pre, y_test))
```

```python
from sklearn.metrics import confusion_matrix
# "{}-{}-{}".format("2019", "02", "03")
mat = confusion_matrix(y_test, pre)
c = ["{}(預測)".format(name) for name in u]
i = ["{}(原本)".format(name) for name in u]
pd.DataFrame(mat, columns=c, index=i)
```
