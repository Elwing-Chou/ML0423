```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
datas = np.random.normal(loc=10, scale=2, size=500000)
sns.distplot(datas)
```

```python
from scipy.stats import skew
datas_2 = np.power(datas, 5)
print("skew:", skew(datas_2))
sns.distplot(datas_2)
```

```python
datas_2 = np.power(2, datas)
print("skew:", skew(datas_2))
sns.distplot(datas_2)
```

```python
datas_2 = np.power(datas, 1/5)
print("skew:", skew(datas_2))
sns.distplot(datas_2)
```
